# Q2 Load the following image and detect faces,eye,smile and write under every detection label Peter ,eye ,smile as arranged.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eyeClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
smileClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_smile.xml')
```

```python
img=cv2.imread("d42473-018-00009-z_15736222.jpg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12,12))
#plt.imshow(gray,cmap='gray')
faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=1,minSize=(20,20))


for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
    face=gray[y:y+h,x:x+w]
    eyes=eyeClassifier.detectMultiScale(face,scaleFactor=1.1,minNeighbors=1,minSize=(80,80))
    for x2,y2,w2,h2 in eyes:
        cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(0,255,0),5)
        print(eyes)
    smiles=smileClassifier.detectMultiScale(face,scaleFactor=1.01,minNeighbors=6,minSize=(70,70))
    for x2,y2,w2,h2 in smiles:
        cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(0,255,0),5)
        print(eyes)

cv2.putText(img, 'eye', (180,300), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,255),1,cv2.LINE_4)
cv2.putText(img, 'eye', (330,300), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,255),1,cv2.LINE_4)
cv2.putText(img, 'smile', (245,400), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,255),1,cv2.LINE_4)
cv2.putText(img, 'peter', (245,430), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,255,255),1,cv2.LINE_4)

       
plt.imshow(img[:,:,::-1])
```

# Q3Run your camera stream and make detection for faces,eye,smile and write under every detection label me ,eye ,smile as arranged. Make circle for eye and others rectangle

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

cap=cv2.VideoCapture(0)

faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eyeClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
smileClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_smile.xml')

while (cap.isOpened()):

    ret,frame=cap.read()

    if (ret):
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=2,minSize=(70,70))
        for x,y,w,h in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),5)
            
        eyes=eyeClassifier.detectMultiScale(face,scaleFactor=1.1,minNeighbors=1,minSize=(80,80))
        for x2,y2,w2,h2 in eyes:
            cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(0,255,0),5)

        smiles=smileClassifier.detectMultiScale(face,scaleFactor=1.01,minNeighbors=6,minSize=(70,70))
        for x2,y2,w2,h2 in smiles:
            cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(0,255,0),5)                        
            
    cv2.imshow("my_image",frame)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
