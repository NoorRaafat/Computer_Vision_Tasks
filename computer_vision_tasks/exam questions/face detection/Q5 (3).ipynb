- load the following image then use haar cascaded detection to detect face, eye and smile ,draw Red rectangle around face and write "Face" below rectangle ,draw Green Circle around eyes and write "eye"below it and draw Red rectangle around smile and write smile below rectangle , then when press "b" apply blur filter with kernel size 9 on just the face but when i press "i" show the image with detected (face,eye,smile)and words below it.when i press "q" stop the program

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("viola.jpg")
plt.imshow(img[:,:,::-1])
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cv2.data.haarcascades
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')
```

```python
faces = classifier.detectMultiScale(gray, scaleFactor = 1.5, minNeighbors = 6, minSize = (50, 50))
```

```python
print(len(faces))
```

```python
for face in faces:
    x, y, w, h = face
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)
    face_roi = gray[y:y+h, x:x+w]
    
    eyes = eye_cascade.detectMultiScale(face_roi, scaleFactor = 1.04, minNeighbors = 5, minSize = (20, 20))
    for eye in eyes:
        x2, y2, w2, h2 = eye
        cv2.rectangle(img, (x+x2, y+y2), (x+x2+w2, y+y2+h2), (0, 255, 0), 4)
        
        
    smiles = smile_cascade.detectMultiScale(face_roi, scaleFactor = 1.1, minNeighbors = 6, minSize = (50, 50))
    for smile in smiles:
        x3, y3, w3, h3 = smile
        cv2.rectangle(img, (x+x3, y+y3), (x+x3+w3, y+y3+h3), (0, 0, 255), 4)
```

```python
print(len(eyes))
print(len(smiles))
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(img[:,:,::-1])
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(img, "Face", (150, 270), font, 1, (0, 0, 255), 5)
cv2.putText(img, "eye", (160, 170), font, 1, (0, 255, 0), 3)
cv2.putText(img, "eye", (240, 170), font, 1, (0, 255, 0), 3)
cv2.putText(img, "smile", (180, 240), font, 1, (0, 0, 255), 3)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(img[:,:,::-1])
```

```python
while True: 
    char = input()
    
    if char == 'b':
        blurred = cv2.blur(img, (9, 9))
        cv2.imshow("Image", blurred)
        
    elif char == 'i':
        cv2.imshow("Image", img)
        
    k = cv2.waitKey(1)
    elif k == 'q':
        break
        
cv2.destroyAllWindows() 
```
