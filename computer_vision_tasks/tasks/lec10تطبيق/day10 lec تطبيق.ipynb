# line

```python
img=np.zeros((500,500,3),np.uint8)
cv2.line(img,(100,0),(200,400),(255,0,0),5)
plt.imshow(img[:,:,::-1])
```

```python
def nothing(x):
    pass

cv2.namedWindow("image")
cv2.createTrackbar("x0","image",0,500,nothing)
cv2.createTrackbar("yo","image",0,500,nothing)
cv2.createTrackbar("x1","image",0,500,nothing)
cv2.createTrackbar("y1","image",0,500,nothing)
cv2.createTrackbar("th","image",1,20,nothing)


while True:
    img_f=np.zeros((500,500,3),np.uint8)
    x0=cv2.getTrackbarPos("x0","image")
    yo=cv2.getTrackbarPos("yo","image")
    x1=cv2.getTrackbarPos("x1","image")
    y1=cv2.getTrackbarPos("y1","image")
    th=cv2.getTrackbarPos("th","image")
    cv2.line(img_f,(x0,yo),(x1,y1),(255,0,0),th)
    cv2.imshow("image",img_f)
    k=cv2.waitKey(1)
    if k== ord('q'):
        break

        

cv2.destroyAllWindows()

```

# rectangle

```python
img=np.zeros((500,500,3),np.uint8)

cv2.rectangle(img,(100,50),(250,250),(0,0,255),10)

plt.imshow(img)


cv2.line(img,(100,50),(250,250),(0,0,255),10)
cv2.line(img,(175,50),(175,250),(0,0,255),10)
cv2.line(img,(100,250),(250,50),(0,0,255),10)
cv2.line(img,(100,150),(250,150),(0,0,255),10)

plt.imshow(img[:,:,::-1])
```

# circle

```python
img=np.zeros((500,500,3),np.uint8)

cv2.circle(img,(250,250),100,(0,0,255),5)

plt.imshow(img)
```
